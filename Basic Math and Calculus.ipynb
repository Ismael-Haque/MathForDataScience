```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def my_function(x):
    return x**2
```

```python
slope_at_2 = derivative_x(my_function, 2, .00001)
slope_at_2
```

## Partial Derivatives

```python
x,y = symbols('x y')
f = 2*x**3 + 3*y**3
ddx_f = diff(f, x)
ddy_f = diff(f, y)
print(ddx_f, ddy_f)
```

```python
plot3d(f)
```

```python
ddx_f.subs(x,1)
```

```python
ddy_f.subs(y,2)
```

## Limit

```python
x, s = symbols('x s')
f = x**2
slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)
slope_f
```

```python
result = limit(slope_f, s, 0)
result
```

## Chain Rule

```python
z = (x**2 + 1)**3 -2
dz_dx = diff(z, x)
dz_dx
```

```python
x, y = symbols('x y')

_y = x**2 + 1
dy_dx = diff(_y)
print(dy_dx)

z = y**3 - 2
dz_dy = diff(z)
print(dz_dy)
```

```python
dz_dx_chain = (dy_dx * dz_dy)
dz_dx_chain
```

```python
dz_dx_chain.subs(y, _y)
```

```python
# no chain rule
dz_dx_no_chain = diff(z.subs(y, _y))
dz_dx_no_chain
```
